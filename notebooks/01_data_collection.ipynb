```python
import yfinance as yf
import pandas as pd

# Download NIFTY 50 data
ticker = "^NSEI"

data = yf.download(ticker, start="2015-01-01", end="2025-01-01")

# Show first rows
data.head()
```

```python
# Properly flatten columns
data.columns = data.columns.map(lambda x: x[0] if isinstance(x, tuple) else x)

# Remove column name labels (Price, Ticker)
data.columns.name = None

# Verify
print(data.columns)
data.head()
```

```python
data.to_csv("../data/nifty50_raw.csv")
```
